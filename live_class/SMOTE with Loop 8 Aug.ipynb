```python
x=winedf.iloc[:,0:-1]
y = winedf.iloc[:,-1]
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
```

```python
xtrainw,ytrainw = sm.fit_resample(x,y)
```

```python
ytrainw
```

```python
ytrainw.value_counts()
```

```python
sns.countplot(ytrainw)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xtrainw,ytrainw)
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(xtrain,ytrain)
```

```python
prd = dtc.predict(xtest)
```

```python
print('Accuracy:: ', accuracy_score(ytest,prd))
confusion_matrix(ytest,prd)
```
